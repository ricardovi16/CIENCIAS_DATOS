###Creacion de dataframe por medio de diccionarios

```python
elementos={
    "Numero atomico":[1,6,47,88],
    "Masa atomica":[1,12,107,226],
    "Familia":["No metal","No metal","Metal","Metal"]
}
elementos
```

```python
tabla_periodica=pd.DataFrame(elementos)
tabla_periodica
```

```python
tabla_periodica=pd.DataFrame(elementos,
                             index=["Hidrogeno","Carbon","Plata","Radio"],
                             columns=["Familia","Numero atomico", "Masa atomica"])
tabla_periodica
```

###Creando un dataframe a partir de un numpy

###Creacion de dataframes con diccionarios


```python
medida_2020={"Ag":2,"Au":5,"Cu":3,"Pt":2}
medida_2021={"Ag":4,"Au":3,"Cu":6,"Pt":1}
medida_2022={"Ag":8,"Au":2,"Cu":15,"Pt":0}
unidades2=pd.DataFrame([medida_2020,medida_2021,medida_2022],
                       index=list(range(2020,2023)))
unidades2
```

```python
medida_2020={"Ag":2,"Au":5,"Cu":3,"Pt":2}
medida_2021={"Ag":4,"Au":3,"Cu":6,"Pt":1}
medida_2022={"Ag":8,"Pb":2,"Cu":15,"Pt":0}
unidades2=pd.DataFrame([medida_2020,medida_2021,medida_2022],
                       index=list(range(2020,2023)))
unidades2
```
